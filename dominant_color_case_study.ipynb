```python
image = cv2.imread('image10.jpg')
#print image
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#print image
```

```python
image = image.reshape((image.shape[0]*image.shape[1],3))
print image
print image.shape
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(image)
```

```python
# initialising the label dictionary
label = [0,0,0]
```

```python
# finding count of data in each cluster
for x in kmeans.labels_:
    label[x] += 1
```

```python
# total number of data points
label_sum = image.shape[0]
#type (label_sum)
```

```python
#converting to percentage
label = [float(x)/label_sum for x in label]
print label
print max(label)
```

```python
for i in range(0,11):
    image_name = 'image' + str(i) + '.jpg'
    # reading a image
    image = cv2.imread(image_name)
    # printing a image which will be in the form of ndarray
    # print image
    # to display width x height x channel
    image.shape

    # to display image
    #plt.figure()
    # setting image axis
    #plt.axis("off")
    # displaying image
    #p = plt.imshow(image)
    print p

    # convertingg 3d image array to 2d
    image = image.reshape((image.shape[0]*image.shape[1],3))
    # print image
    print image.shape

    # declaring object of KMeans
    kmeans = KMeans(n_clusters=5, random_state=0).fit(image)

    # initialising the label dictionary
    label = [0,0,0,0,0]

    # finding count of data in each cluster
    for x in kmeans.labels_:
        label[x] += 1

    # total number of data points
    label_sum = image.shape[0]

    #converting to percentage
    label = [float(x)/label_sum for x in label]
    print label

    print kmeans.cluster_centers_

    centroids = [x.astype('uint8').tolist() for x in kmeans.cluster_centers_]

    clr = [webcolors.rgb_to_hex(c) for c in centroids]
    print 'colour: '+ str(clr)

    clr_name = []
    for c in clr:
        clr_name.append(webcolors.rgb_to_name(c))
    print clr_name
    
    # ploting pie chart
    plt.pie(label, colors=clr, shadow=True)
    plt.axis('equal')
    plt.show()
```
