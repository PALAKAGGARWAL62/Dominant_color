```python
# for image recogonition 
import cv2
# for dataframes
import pandas as pd
# to print graphs
import matplotlib.pyplot as plt
# perform numeric operations
import numpy as np
# to print graphs on the same window
%matplotlib inline
# to make matplotlib coding easy
import seaborn as sns
# object of sns
sns.set()
```

```python
# reading image
image = cv2.imread('C:\Users\hp\Documents\ml practice\clorify\Dominant Color Data Set\image 10.jpg')
#print image
```

```python
# converting image pixel intensity values from BGR to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#print image
```

```python
# to print image structure width x height x channel
image.shape
```

```python
# to print image
plt.figure()
plt.axis("off")
p = plt.imshow(image)
print p
```

```python
# converting image 2d array to 1d array
image = image.reshape((image.shape[0]*image.shape[1],3))
print image
print image.shape
```

```python
# importing KMeans algorithm from sklearn library
from sklearn.cluster import KMeans
```

```python
# making object of KMeans and fitting image to model
kmeans = KMeans(n_clusters=3, random_state=0).fit(image)
```

```python
# initialising the label dictionary
label = [0,0,0]
```

```python
# finding count of data in each cluster
for x in kmeans.labels_:
    label[x] += 1
```

```python
# total number of data points
label_sum = image.shape[0]
#type (label_sum)
```

```python
#converting to percentage
label = [(float(x)/label_sum)*100 for x in label]
print label
m = max(label)
print m
```

```python
# printing colour intensity at each label centroid
print kmeans.cluster_centers_
```

```python
# round of color values
centroids = [x.astype('uint8').tolist() for x in kmeans.cluster_centers_]
print centroids
```

```python
# to convert RGB codes to color hex codes
import webcolors
```

```python
# converting RGB codes to color hex codes
clr = [webcolors.rgb_to_hex(c) for c in centroids]
print clr
```

```python
# ploting pie chart
plt.pie(label, colors=clr, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.show()
```

```python
# data frame for dominant color in each image
domc = pd.DataFrame({
    'Image name': [],
    'Dominant colour percentage': [],
    'Color': []
})
```

```python
# list to store dominant color realated data
data = []
# fetching images in the loop
for i in range(1,11):
    
    print '****************************************************************************** '
    print '****************************** IMAGE %s *************************************** ' %i
    print '****************************************************************************** '
    
    image_name = 'C:\Users\hp\Documents\ml practice\clorify\Dominant Color Data Set\image ' + str(i) + '.jpg'
    
    # reading a image
    image = cv2.imread(image_name)
    
    # covert image pixel values from BGR to RBG
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # to display width x height x channel
    print 'Image shape: '+str(image.shape)
    
    '''
    # to display image
    plt.figure(i)
    # setting image axis
    plt.axis("off")
    # displaying image
    plt.imshow(image)
    '''
    
    # converting 3d image array to 1d
    image = image.reshape((image.shape[0]*image.shape[1],3))
    # print image
    print 'Shape of image after image reshape: '+str(image.shape)

    # declaring object of KMeans
    kmeans = KMeans(n_clusters=3, random_state=0).fit(image)

    # initialising the label dictionary
    label = [0,0,0]

    # finding count of data in each cluster
    for x in kmeans.labels_:
        label[x] += 1

    # total number of data points
    label_sum = image.shape[0]

    #converting to percentage
    label = [round((float(x)/label_sum)*100, 2) for x in label]
    m = max(label)
    y = label.index(m)
    print 'Percentage of different colours: '+str(label)

    # print kmeans.cluster_centers_

    # color values of each centeroid pixels of different clusters
    centroids = [x.astype('uint8').tolist() for x in kmeans.cluster_centers_]
    print centroids
    
    # covert pixel RGB color intensity values to hex code
    clr = [webcolors.rgb_to_hex(c) for c in centroids]
    print 'colours hex codes: '+ str(clr)
    dc = clr[y]

    # ploting pie chart
    plt.pie(label, colors=clr, shadow=True, autopct='%1.1f%%')
    plt.axis('equal')
    plt.show()
    
    # appending data to list
    l = []
    l.append('Image'+str(i))
    l.append(m)
    l.append(dc)
    
    # appending list all the image details to a list
    data.append(l)
    
    # appending data to dataframe
    domc = domc.append(pd.DataFrame({
        'Image name' : [data[i-1][0]],
        'Dominant colour percentage' : [data[i-1][1]],
        'Color' : [data[i-1][2]]
    }))
```
